```
import requests
from bs4 import BeautifulSoup

def url_to_soup(url):
    req = requests.get(url)
    return BeautifulSoup(req.content, 'html.parser')
#url_to_soup('https://www.nankankeiba.com/race_info/2021032420190301.do')
```

```
def horse_page_link(url):
    soup = url_to_soup(url)
    link_list = [x.get('href') for x in soup.find_all('a', class_='tx-mid tx-low') ]
    return link_list

#horse_page_link('https://www.nankankeiba.com/race_info/2021032420190301.do')

```
