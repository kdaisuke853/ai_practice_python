```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
target_url = 'https://www.nikkei.com/nkd/company/?scode=6172'
r = requests.get(target_url)
```

```python
soup = BeautifulSoup(r.text,'lxml')
```

```python
val = soup.find('dt', attrs={ 'class': 'm-stockPriceElm_title' })
```

```python
val2 = str(val)
```

```python
val2
```

```python
a = re.search(r'([0-9]*:[0-9]*)', val2)
```
